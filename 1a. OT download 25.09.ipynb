## Create combined output files

```python
import pandas as pd
import os

directory = './association_by_datasource_direct'
parquet_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.parquet')]

df_combined = pd.concat([pd.read_parquet(f) for f in parquet_files])
df_combined.to_parquet('./OT_Raw/direct.parquet', index=False)

###

directory = './target'
parquet_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.parquet')]

df_combined = pd.concat([pd.read_parquet(f) for f in parquet_files])
df_combined.to_parquet('./OT_Raw/targets.parquet')

###

directory = './disease'
parquet_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.parquet')]

df_combined = pd.concat([pd.read_parquet(f) for f in parquet_files])
df_combined.to_parquet('./OT_Raw/diseases.parquet')

###

directory = './drug_molecule'
parquet_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.parquet')]

df_combined = pd.concat([pd.read_parquet(f) for f in parquet_files])
df_combined.to_parquet('./OT_Raw/molecule.parquet')

###

directory = './drug_mechanism_of_action'
parquet_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.parquet')]

df_combined = pd.concat([pd.read_parquet(f) for f in parquet_files])
df_combined.to_parquet('./OT_Raw/mechanismOfAction.parquet')

###

directory = './drug_indication'
parquet_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.parquet')]

df_combined = pd.concat([pd.read_parquet(f) for f in parquet_files])
df_combined.to_parquet('./OT_Raw/indication.parquet')

###

directory = './evidence/sourceId=chembl'
parquet_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.parquet')]

df_combined = pd.concat([pd.read_parquet(f) for f in parquet_files])
df_combined.to_parquet('./OT_Raw/chembl.parquet')
```
